# Data file

The data file for revised Simos' method is a [.csv](https://en.m.wikipedia.org/wiki/Comma-separated_values) text file that contains:

- Criteria grouped in categories from least important to most important; criteria considered equal are in the same category.
- Gaps between successive categories are indicated by `white` cards.
- Ratio between the importance of the most and least important category.

Examples data files are given in `data\` folder. Data files can be obtained with a [spreadsheet](https://en.m.wikipedia.org/wiki/Spreadsheet).

# Problem solving

The weighting problem is:

Given a set of ordered cards (from least to most important), and the ratio of the importance between most and least important, obtain the weights of each criterion.

```python
set_cards = "../../data/cards_subsets_3.csv"

# The last subset is `z` more important than the 1st one
z = 5

df = Simos_revised.criteria_weights_Simos(set_cards, z)
```

# Results

The result is a list of criteria (in alphabetical order) and their weights.

```python
print(f"\nResults for {set_cards} with z = {z}:\n")
print("Sorted by criteria name")
df.sort_values(by=['Criteria'])
```

The result may be is a list of criteria ordered in function of their weights.

```python
print("Sorted weight values")
df.sort_values(by=['Weight'])
```
